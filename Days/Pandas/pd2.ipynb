# Filtering and Sorting 

When there is a large dataset, then the process of fetching the necessary data only at a time is known as filtering

Arranging data into particular format is sorting 

```python
df.sample(10)
```

# Filtering 

```python
# Task 1 -> Display the records of those students whose last name is Gupta

filt = (df["Last_Name"] == "Gupta")
```

```python
df.loc[filt]
```

```python
# Task 2 -> From the above task, print only First_Name, Roll_Number and Total_Marks
df.loc[filt, ("First_Name", "Roll_Number", "Total_Marks")]
```

```python
# Task 3 -> Display all others record than the defined filter 
```

```python
df.loc[~filt, ("First_Name", "Roll_Number", "Total_Marks")]
```

```python
L_n_lis = ["Gupta", "Thakur"]
filt = df["Last_Name"].isin(L_n_lis)
```

```python
df.loc[filt]
```

```python
df.loc[filt, ("First_Name", "Last_Name", "Total_Marks")]
```
