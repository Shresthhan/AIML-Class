```python
import pandas as pd
```

```python
# Mathematical Formulation for Outlier Detection
'''
1 -> Compute Q1 and Q3
2 -> Inter Quartile Range(IQR) = Q3 - Q1
3 -> Compute the upper and lower bound -> Lower = Q1 - 1.5 * IQR -> Upper = Q3 + 1.5 * IQR
'''
```

```python
df = pd.read_csv('/content/outlier_dataset.csv')
```

```python
df.sample(5)
```

```python
def IQR_bounds(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  print(f"For {column} \n Upper = {upper} \n Lower = {lower}")
  return lower, upper

def remove_outliers(data, column):
  lower, upper = IQR_bounds(data, column)
  filt = ((data[column] >= lower) & (data[column] <= upper))
  return data[filt]
```

```python
df_cleaned = (
    df
    .pipe(remove_outliers, column = "Salary")
    .pipe(remove_outliers, column = "Age")
)
print("Original Dataset: \n", df)
print("Cleaned Dataset: \n", df_cleaned)
```

```python
df_cleaned.to_csv("day2_CLean_Data.csv", index = False)
```
