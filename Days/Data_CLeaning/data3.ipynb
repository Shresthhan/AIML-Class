```python
import pandas as pd
```

```python
df = pd.read_csv('sample_dirty_dataset.csv')
```

```python
df
```

```python
# Handiling Missing Values
'''
1 -> Remove Outliers
'''
def remove_outliers_IQR(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
df_clean = (
    df
    .pipe(remove_outliers_IQR, column = "Age")
    .pipe(remove_outliers_IQR, column = "Salary")
)
```

```python
df_clean
```

```python
# Fill the Unknown Values -> Strings
df_clean["Department"].fillna("Unknown", inplace=True)
```

```python
df_clean
```

```python
# Remove Duplicates
df_clean.drop_duplicates(inplace=True)
```

```python
# Making Data Consistent
df_clean["Gender"] = df["Gender"].replace({"Male": "M", "Female": "F"})
```
